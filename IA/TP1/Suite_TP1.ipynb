### 12. Triez le DataFrame par le nombre de produits dans chaque transaction, de manière décroissante, puis affichez les premières lignes

```python
df_trie = df.sort_values(by='nb produits par transaction', ascending=False)
df_trie
```

### 13. Affichez les produits les plus fréquemment achetés par transaction (i.e. le plus fréquent dans une transaction)

```python
produits = df.apply(lambda row: row.dropna().values, axis=1).explode()

produits.value_counts()
```

# Exercice 4 : Suppression d’éléments dans le DataFrame

## 1. Supprimez la colonne Nombre de produits ajoutée dans l’exercice précédent

```python
df = df.drop(columns=['nb produits par transaction'])
df.head()
```

## 2. Supprimez les colonnes vides, c’est-à-dire toutes les colonnes qui ne contiennent que des valeurs manquantes (NaN)

```python
df = df.dropna(axis=1, how='all')
df.head()
```

## 5. Supprimez les lignes qui contiennent au moins une valeur manquante

```python
df_temp = df.dropna()
df_temp.head()
```

## 4. Supprimez les transactions identiques (doublons)

```python
df.drop_duplicates()
```

## 5. Supprimez toutes les transactions qui contiennent le produit avec l'identifiant 42

```python
df = df[~df.apply(lambda row: 42 in row.values, axis=1)]
df
```

## 6. Supprimez les transactions contenant moins de 3 produits

```python
df = df[df.count(axis=1) >= 3]
df
```
