```python
import random
import time
import matplotlib.pyplot as plt
```

```python
def createListRandom(n):
    liste = []
    for i in range(n):
        liste.append(i)
        random.shuffle(liste)
    return liste

def measure_time(function):
    def measure_fun(*args):
        t = time.perf_counter()
        res = function(*args)
        return time.perf_counter() - t, res
    return measure_fun
```

# Tri par insertion

```python
@measure_time
def tri_par_insertion(liste):
    if len(liste) == 0 or len(liste) == 1:
        return liste
    listeTrie = []
    listeTrie.append(liste[0])
    for i in range(1, len(liste)):
        for j in range(len(listeTrie)):
            if liste[i] < listeTrie[0]:
                listeTrie.insert(0, liste[i])
                break
            if liste[i] < listeTrie[j]:
                listeTrie.insert(j, liste[i])
                break
            if j == len(listeTrie) - 1:
                listeTrie.append(liste[i])
                break
    return listeTrie
```

## Mesure

```python
times = []
nbElements = []
for i in range(1_000, 10_000, 1_000):
    times.append(tri_par_insertion(createListRandom(i)))
    nbElements.append(i)

print(times)
plt.plot(nbElements, times)
plt.title('Graphique simple')
plt.xlabel('Nb Elements')
plt.ylabel('Temps')
plt.show()
```

# Tri à bulle

```python
@measure_time
def tri_a_bulle(liste):
    n = len(liste)
    for i in range(n):
        for j in range(0, n-i-1):
            if liste[j] > liste[j+1]:
                liste[j], liste[j+1] = liste[j+1], liste[j]
    return liste
```

## Mesure

```python
times = []
nbElements = []
for i in range(1_000, 10_000, 1_000):
    times.append(tri_a_bulle(createListRandom(i)))
    nbElements.append(i)

print(times)
plt.plot(nbElements, times)
plt.title('Graphique simple')
plt.xlabel('Nb Elements')
plt.ylabel('Temps')
plt.show()
```
